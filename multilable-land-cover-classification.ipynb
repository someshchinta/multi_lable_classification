```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D,Conv3D, MaxPooling3D
from keras.utils import to_categorical
from keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tqdm import tqdm
%matplotlib inline
```

```python
#!wget https://www.dropbox.com/s/u83ae1efaah2w9o/UCMercedLanduse.zip
#!unzip UCMercedLanduse.zip
#!unzip UCMercedLanduse.zip -d "./input/data_images"
```

```python
train = pd.read_csv("../input/labels2/labels/labels/multilabels.txt", sep="\t")   # reading the csv file
train.head()      # printing first five rows of the file
#train.keys()
```

```python
train_image = []
for i in tqdm(range(train.shape[0])):
    #print(i)
    img = image.load_img('../input/images2/images/'+train['IMAGE\LABEL'][i]+'.tif',target_size=(400,400,3))
    img = image.img_to_array(img)
    #img = image.img_to_array(img)
    img = img/255
    train_image.append(img)
X = np.array(train_image)
print(X.size)
```

```python
import matplotlib.pyplot as plt
plt.imshow(X[2])
```

```python
print(X.shape)
y = np.array(train.drop(['IMAGE\LABEL'],axis=1))
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=12, test_size=0.15)
```

```python
from keras.layers.normalization import BatchNormalization

model = Sequential()
model.add(Conv2D(28, (3, 3), activation='relu', input_shape=(400 , 400 , 3)))
model.add(BatchNormalization(epsilon=1e-06,  momentum=0.9, weights=None))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
 
model.add(Conv2D(56, (3, 3), activation='relu'))
model.add(BatchNormalization(epsilon=1e-06,  momentum=0.9, weights=None))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
 
model.add(Conv2D(56, (3, 3), activation='relu'))
model.add(BatchNormalization(epsilon=1e-06, momentum=0.9, weights=None))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))







model.add(Flatten())

model.add(Dense(512, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.25))


model.add(Dense(17, activation='softmax'))








```

```python
model.summary()
```

```python
from keras import optimizers
#rms = optimizers.RMSprop(lr=0.0001, rho=0.9, epsilon=None, decay=0.0)
#model.compile(optimizer=rms, loss='categorical_crossentropy', metrics=['accuracy',keras_metrics.precision(), keras_metrics.recall()])
opt=optimizers.Adam(lr=0.001, decay=2e-3)
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history=model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test),batch_size=64)
```

```python
accuracy = model.evaluate(X_test, y_test)
#print("Score: ", score)
print("Accuracy: ", accuracy)
#print("Precision: ", precision)
#print("Recall: ", recall)
```

```python
# Loss Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['loss'],'r',linewidth=3.0)
plt.plot(history.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
 
# Accuracy Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['accuracy'],'r',linewidth=3.0)
plt.plot(history.history['val_accuracy'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```
